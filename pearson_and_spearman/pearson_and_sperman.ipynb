# Pearson and Spearman correlation
Pearson  關心變數間的*線性相關*
  
$Pearson(x, y) = \frac{\Sigma{(x-\hat{x})(y-\hat{y})}}{\Sigma{(x-\hat{x})^2} \Sigma{(y-\hat{y})^2}}$    

Spearman 關心變數間的 *單調相關*

$Spearman(x, y) = Pearson(ro_rank(x), to_rank(y))$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
def to_rank_arr(arr):
  return arr.argsort().argsort()

def pearson(x, y):
  return sum((x - x.mean())*(y - y.mean())) \
      /np.sqrt((sum((x - x.mean())**2)*sum((y - y.mean())**2)))

def spearman(x, y):
  return pearson(to_rank_arr(x), to_rank_arr(y))
```

```python
x = np.array([1, 2, 3, 4, 6, 5, 9])
y = np.array([20, 30, 40, 6, 7, 4, 5])
plt.scatter(x, y)
print(f"pearson={pearson(x, y)}, spearman=: {spearman(x, y)}")
```

```python
x = np.arange(0, 100)
y = np.linspace(0, 1, 100)
plt.scatter(x, y, alpha=0.8)
print(f"pearson={pearson(x, y)}, spearman=: {spearman(x, y)}")
```

```python
x = np.arange(0, 100)
y = np.linspace(0, 1, 100) + np.random.normal(0, 0.3, 100)
plt.scatter(x, y, alpha=0.8)
print(f"pearson={pearson(x, y)}, spearman=: {spearman(x, y)}")
```
